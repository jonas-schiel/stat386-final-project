```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

1948 Table

```python
#Because of inconsistencies in how the data is formatted we will actually need to scrape and clean tables one at a time

df_1948 = []

url = f"https://www.presidency.ucsb.edu/statistics/elections/1948"
    #We use requests to get the URL
webpage = requests.get(url)
    #We then use the beautifulsoup function to use an html parser to scan the dev text of the website
soup = BeautifulSoup(webpage.text, "html.parser")
    #the .find  function will look through soup to find the word table. Since there is only one table in the dev text we don't need to worry
    #about it the wrong table being tagged
table = soup.find("table")

    #We then use read_html from pandas to scrape the tagged section of the dev code. Str() converts the contents to string type. [0] returns a df
dataframe = pd.read_html(str(table))[0]
    #this adds year as a variable in the data frame
dataframe["year"] = 1948

#create the dataframe using append and concat
df_1948.append(dataframe)
df_1948 = pd.concat(df_1948, ignore_index = True)

#Rename columns to the correct varaible names and add president elect and party
df_1948["State"] = df_1948[0]
df_1948["Total Votes"] = df_1948[1]
df_1948["Republican Votes"] = df_1948[5]
df_1948["Republican Percentage"] = df_1948[6]
df_1948["Republican Electoral"] = df_1948[7]
df_1948["Democratic Votes"] = df_1948[2]
df_1948["Democratic Percentage"] = df_1948[3]
df_1948["Democratic Electoral"] = df_1948[4]
df_1948["President Elect"] = str("Harry S Truman")
df_1948["Winning Party"] = str("D")

#Remove all duplicate columns
df_1948 = df_1948[["State", "Total Votes", "Republican Votes", 
                  "Republican Percentage", "Republican Electoral", "Democratic Votes",
                  "Democratic Percentage", "Democratic Electoral", "President Elect", "Winning Party", "year"]]

#Drop the first 11 rows of the data
df_1948 = df_1948.iloc[12:].reset_index(drop = True)
#Drop every row below 48
df_1948 = df_1948.iloc[:48].reset_index(drop = True)

#Check top and bottom of data 
pd.concat([df_1948.head(), df_1948.tail()])
```
